```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tsv_names = pd.read_csv(r"..\data.tsv", sep='\t')
```

```python
tsv_names
```

```python
tsv_names['titleType'].unique()
```

```python
movies = tsv_names[(tsv_names['titleType'] == 'movie') | (tsv_names['titleType'] == 'tvMovie')]
```

```python
movies
```

```python
movies70s = movies[movies['startYear'].str.startswith('197')]
movies80s = movies[movies['startYear'].str.startswith('198')]
movies90s = movies[movies['startYear'].str.startswith('199')]
movies00s = movies[movies['startYear'].str.startswith('200')]
movies10s = movies[movies['startYear'].str.startswith('201')]
movies20s = movies[movies['startYear'].str.startswith('202')]
```

```python
movies = pd.concat([movies70s,movies80s,movies90s,movies00s,movies10s,movies20s])
```
